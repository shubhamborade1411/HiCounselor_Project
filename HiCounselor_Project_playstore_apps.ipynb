## Dealing with missing values in Rating Feature

```python
from sklearn.impute import KNNImputer 
```

```python
imputer = KNNImputer(n_neighbors=12)
df['Rating']=imputer.fit_transform(df[['Rating']])
```

```python
df.isna().sum()
```

## Dealing with duplicated values

```python
df.duplicated('App')
```

```python
df[df.duplicated('App')==True]
```

```python
df[df['App']=='Quick PDF Scanner + OCR FREE']
```
